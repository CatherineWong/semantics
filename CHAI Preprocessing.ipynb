```python
with open('landau_jackendoff_prepositions.json') as f:
    spatial_prepositions = json.load(f)

print(spatial_prepositions)
```

```python
import sys
import csv

csv.field_size_limit(sys.maxsize)

chai_sentences = []
with open('data/chai/text2traj_instructions_2018_emnlp_instructions.csv', 'r') as f:
    csv_reader = csv.DictReader(f)
    for row in csv_reader:
        chai_sentences += row['inst1'].split('\n')
        chai_sentences += row['inst2'].split('\n')
    
        
    
```

```python
preposition_samples = {}
for preposition in spatial_prepositions:
    attested_sentences = []
    for text in chai_sentences:
        if ' ' + preposition + ' ' in text:
            # Extract sentence
            sentences = [sentence for sentence in text.split('.') if ' ' +preposition + ' '  in sentence]
            attested_sentences += sentences
            
    preposition_samples[preposition] = attested_sentences
```

```python
for preposition in sorted(preposition_samples.keys()):
    if len(preposition_samples[preposition]) > 0:
        print(preposition.upper())
        attested = sorted(preposition_samples[preposition], key=lambda s:len(s))
        for sent in attested[:5]:
            sent = sent.replace(preposition, '*' + preposition + '*')
            print('- ' + sent)
        print('\n\n')
```

```python
unattested = [p for p in preposition_samples.keys() if len(preposition_samples[p]) > 0]
print(unattested)
```

```python
import nltk
from nltk import word_tokenize, pos_tag
all_verbs = defaultdict(list)

for full_text in chai_sentences:
    for text in full_text.split('.'):
        tagged = pos_tag(word_tokenize(text.lower()))
        verbs = [tag[0] for tag in tagged if 'VB' == tag[1]]
        
        for verb in verbs:
            all_verbs[verb].append(text.replace(verb, '*' +verb +'*'))

for verb in sorted(all_verbs.keys()):
    print(verb.upper())
    attested = sorted(all_verbs[verb], key=lambda s:len(s))
    for sent in attested[:5]:
        print('- ' + sent)
    print('\n\n')
```
