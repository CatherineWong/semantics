```python
import json
```

```python
# Reference: Jackendoff / Landau 1993
spatial_prepositions = [
    'about', 'above', 'across', 'after', 'against', 'along', 'alongside', 'amid', 'amidst', 'among', 'amongst',
    'around', 'at', 'atop', 'behind', 'below', 'beneath', 'beside', 'between', 'betwixt',
    'beyond', 'by', 'down', 'from', 'in', 'inside', 'into', 'near', 'nearby', 'off', 'on',
    'onto', 'opposite', 'out', 'outside', 'over', 'past', 'through', 'throughout', 'to', 'toward', 'under',
    'underneath', 'up', 'upon', 'via', 'with', 'within', 'without',
    'far', 'back of', 'between', 'front of','top of', 'left of', 'right of', 'side of',
    'afterward', 'afterwards', 'away', 'back', 'backward', 'downward', 'east', 'forward',
    'here', 'inward', 'left', 'north', 'outward', 'right', 'sideways', 'south', 'there',
    'together', 'upward', 'west', 'round'
]
with open('landau_jackendoff_prepositions.json', 'w') as f:
    json.dump(spatial_prepositions, f)
```

```python
with open('landau_jackendoff_prepositions.json') as f:
    spatial_prepositions = json.load(f)
```

```python
all_splits = {}
for split in ['train', 'test', 'dev']:
        raw_json= [json.loads(line) for line in open('data/touchdown/%s.json' % split, 'r')]
        all_splits[split] = raw_json
        
```

```python
from collections import defaultdict
td_strings = defaultdict(list)

for split in ['train', 'test', 'dev']:
    for example in all_splits[split]:
        td_strings['full_text'].append(example['full_text'])
        td_strings['td_location_text'].append(example['td_location_text'])
        td_strings['navigation_text'].append(example['navigation_text'])
```

```python
preposition_samples = {}
for preposition in spatial_prepositions:
    attested_sentences = []
    for text in td_strings['td_location_text']:
        if ' ' + preposition + ' ' in text:
            # Extract sentence
            sentences = [sentence for sentence in text.split('.') if preposition in sentence]
            attested_sentences += sentences
            
    preposition_samples[preposition] = attested_sentences
    
    
```

```python
for preposition in sorted(preposition_samples.keys()):
    print(preposition.upper())
    attested = sorted(preposition_samples[preposition], key=lambda s:len(s))
    for sent in attested[:5]:
        sent = sent.replace(preposition, '*' + preposition + '*')
        print('- ' + sent)
    print('\n\n')
```

```python
unattested = [p for p in preposition_samples.keys() if len(preposition_samples[p]) == 0]
print(unattested)
```

```python
import nltk
from nltk import word_tokenize, pos_tag
all_verbs = defaultdict(list)

for full_text in td_strings['full_text']:
    for text in full_text.split('.'):
        tagged = pos_tag(word_tokenize(text.lower()))
        verbs = [tag[0] for tag in tagged if 'VB' == tag[1]]
        
        for verb in verbs:
            all_verbs[verb].append(text.replace(verb, '*' +verb +'*'))
```

```python
for verb in sorted(all_verbs.keys()):
    print(verb.upper())
    attested = sorted(all_verbs[verb], key=lambda s:len(s))
    for sent in attested[:5]:
        print('- ' + sent)
    print('\n\n')
```
